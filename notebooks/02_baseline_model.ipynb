```python
# Bibliotheken importieren
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report

# Vorverarbeitete Daten laden
X_train = pd.read_csv('../data/X_train.csv')
X_test = pd.read_csv('../data/X_test.csv')
y_train = pd.read_csv('../data/y_train.csv').squeeze()
y_test = pd.read_csv('../data/y_test.csv').squeeze()
sex_test = pd.read_csv('../data/sex_test.csv').squeeze()

print("Daten geladen.")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
```

```python
# Random Forest trainieren
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

print("Modell trainiert.")
```

```python
# Modell evaluieren
y_pred = rf_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("---")
print(classification_report(y_test, y_pred))
```

```python
# Vorhersagen nach Geschlecht aufteilen
y_pred_male = y_pred[sex_test == 1]
y_pred_female = y_pred[sex_test == 0]

y_test_male = y_test[sex_test == 1]
y_test_female = y_test[sex_test == 0]

print("Männer — positive Vorhersagerate:", y_pred_male.mean().round(4))
print("Frauen — positive Vorhersagerate:", y_pred_female.mean().round(4))
print("---")
print("Männer — Accuracy:", round(accuracy_score(y_test_male, y_pred_male), 4))
print("Frauen — Accuracy:", round(accuracy_score(y_test_female, y_pred_female), 4))
```

```python
# Ergebnisse als DataFrame speichern
results_baseline = pd.DataFrame({
    'Metrik': ['Positive Vorhersagerate', 'Accuracy'],
    'Männer': [0.27, 0.805],
    'Frauen': [0.0878, 0.9243]
})

results_baseline.to_csv('../results/tables/baseline_results.csv', index=False)
print("Ergebnisse gespeichert.")
```
